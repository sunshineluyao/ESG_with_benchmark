```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Jun= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Jun.csv")
df_Jun.head()
```

```
df_Jun=df_Jun.dropna()
df_Jun.head()
```

```
### formating data so that it can be calculated
df_Jun[['Size','Common Shares Outstdg','MktVal Co']]=df_Jun[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Jun.head()
```

```
Conds=[
       df_Jun['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Jun['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Jun['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Jun['ESG_Index'] =np.select(Conds, Choices)
df_Jun.head()
```

```
df_Jun[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Jun[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Jun.head()
```

```
### Generate the new variable:BM_Ratio
df_Jun['BM_Ratio'] = df_Jun['Book Value Per Shr Tangble']*df_Jun['Common Shares Outstdg']/df_Jun['MktVal Co']
df_Jun.head()
```

```
df_Jun.count()
```

```
### Generate the categories by quintile cut of size
df_Jun['Size_Index']=pd.qcut(df_Jun['Size'],5,labels=np.arange(1,6,1))
df_Jun.head()
```

```
df_Jun.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Jun['BM_Ratio_Index'] = df_Jun.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Jun.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Jun['Momentum_Index'] = df_Jun.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Jun.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Jun_Overview=df_Jun[['Symbol','Name','Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Jun_Overview.head()
```

```
df_Jun_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=True).size()
```

```
df_Jun_Overview_Benchmark = df_Jun_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Jun_Overview_Benchmark=df_Jun_Overview_Benchmark.rename(columns={"Monthly Return":"Benchmark_Return"})
df_Jun_Overview_Benchmark.head()
```

```
df_Jun_AR = pd.merge(df_Jun_Overview, df_Jun_Overview_Benchmark,  how='left', left_on=['Size_Index','BM_Ratio_Index','Momentum_Index'], right_on = ['Size_Index','BM_Ratio_Index','Momentum_Index'])
df_Jun_AR.head()
```

```
df_Jun_AR['Abnormal_Return']= df_Jun_AR['Monthly Return']-df_Jun_AR['Benchmark_Return']
df_Jun_AR.head()
```

```
###unpivot the table
df_Jun_AR['Month'] = "2020-06"
df_Jun_AR.head()
```
