```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Sep= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Sep.csv")
df_Sep.head()
```

```
df_Sep=df_Sep.dropna()
```

```
df_Sep=df_Sep.set_index('Symbol',drop=True)
df_Sep.head()
```

```
df_Sep.index.name=None
df_Sep.head()
```

```
### formating data so that it can be calculated
df_Sep[['Size','Common Shares Outstdg','MktVal Co']]=df_Sep[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Sep.head()
```

```
Conds=[
       df_Sep['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Sep['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Sep['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Sep['ESG_Index'] =np.select(Conds, Choices)
df_Sep.head()
```

```
df_Sep[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Sep[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Sep.head()
```

```
### Generate the new variable:BM_Ratio
df_Sep['BM_Ratio'] = df_Sep['Book Value Per Shr Tangble']*df_Sep['Common Shares Outstdg']/df_Sep['MktVal Co']
df_Sep.head()
```

```
df_Sep.count()
```

```
### Generate the categories by quintile cut of size
df_Sep['Size_Index']=pd.qcut(df_Sep['Size'],5,labels=np.arange(1,6,1))
df_Sep.head()
```

```
df_Sep.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Sep['BM_Ratio_Index'] = df_Sep.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Sep.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Sep['Momentum_Index'] = df_Sep.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Sep.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Sep_Overview=df_Sep[['Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Sep_Overview.head()
```

```
df_Sep_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).size()
```

```
df_Sep_Overview_Benchmark = df_Sep_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Sep_Overview_Benchmark.head()
```

```
df_Sep_Overview_Leader = df_Sep_Overview[df_Sep_Overview['ESG_Index']=='Leader']
df_Sep_Overview_Laggard = df_Sep_Overview[df_Sep_Overview['ESG_Index']=='Laggard']
df_Sep_Overview_Average = df_Sep_Overview[df_Sep_Overview['ESG_Index']=='Average']
df_Sep_Overview_Leader.head()
```

```
df_Sep_Overveiw_Leader_Mean = df_Sep_Overview_Leader.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Sep_Overveiw_Laggard_Mean = df_Sep_Overview_Laggard.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Sep_Overveiw_Average_Mean = df_Sep_Overview_Average.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Sep_Overveiw_Leader_Mean.head()
```

```
df_Sep_Overview_Benchmark['Leader Monthly Return']= df_Sep_Overveiw_Leader_Mean['Monthly Return']
df_Sep_Overview_Benchmark['Laggard Monthly Return']= df_Sep_Overveiw_Laggard_Mean['Monthly Return']
df_Sep_Overview_Benchmark['Average Monthly Return']= df_Sep_Overveiw_Average_Mean['Monthly Return']
df_Sep_Overview_Benchmark.head()
```

```
df_Sep_Overview_Benchmark['Leader Abnormal Return']=df_Sep_Overview_Benchmark['Leader Monthly Return']-df_Sep_Overview_Benchmark['Monthly Return']
df_Sep_Overview_Benchmark['Laggard Abnormal Return']=df_Sep_Overview_Benchmark['Laggard Monthly Return']-df_Sep_Overview_Benchmark['Monthly Return']
df_Sep_Overview_Benchmark['Average Abnormal Return']=df_Sep_Overview_Benchmark['Average Monthly Return']-df_Sep_Overview_Benchmark['Monthly Return']
df_Sep_Overview_Benchmark.head()
```

```
###unpivot the table
df_Sep_AR = df_Sep_Overview_Benchmark.melt(id_vars=['Size_Index', 'BM_Ratio_Index',	'Momentum_Index'], value_vars=['Leader Abnormal Return', 'Laggard Abnormal Return','Average Abnormal Return'],var_name='ESG Rating', value_name='Abnormal Return')
df_Sep_AR=df_Sep_AR.dropna()
df_Sep_AR.head()
```

```
df_Sep_AR['ESG Rating']=df_Sep_AR['ESG Rating'].apply(lambda x:'Leader' if x=='Leader Abnormal Return' else 'Laggard' if x=='Laggard Abnormal Return' else 'Average' )
df_Sep_AR.head()
```

```
df_Sep_AR['Month']='Sep'
df_Sep_AR.head()
```
