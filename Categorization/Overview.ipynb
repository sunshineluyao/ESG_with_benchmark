```
df_Overview=df_Overview_Jan.append([df_Overview_Feb,df_Overview_Mar,df_Overview_Apr,df_Overview_May,df_Overview_Jun,df_Overview_Jul,df_Overview_Aug ])
df_Overview.head()
```

```
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Box(
    y=df_Overview['Abnormal Return'],x=df_Overview['Month'],
    name='Abnormal Return by Month',
    boxmean=True # represent mean
))
fig.update_layout(
    title='Abnormal Return by Month')
```
