```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
df_Overview_Jan= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Jan_AR.csv",index_col='Unnamed: 0')
df_Overview_Feb= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Feb_AR.csv",index_col='Unnamed: 0')
df_Overview_Mar= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Mar_AR.csv",index_col='Unnamed: 0')
df_Overview_Apr= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Apr_AR.csv",index_col='Unnamed: 0')
df_Overview_May= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/May_AR.csv",index_col='Unnamed: 0')
df_Overview_Jun= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Jun_AR.csv",index_col='Unnamed: 0')
df_Overview_Jul= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Jul_AR.csv",index_col='Unnamed: 0')
df_Overview_Aug= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Aug_AR.csv",index_col='Unnamed: 0')
df_Overview_Sep= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/Categorization/Sep_AR.csv",index_col='Unnamed: 0')
```

```
df_Overview=df_Overview_Jan.append([df_Overview_Feb,df_Overview_Mar,df_Overview_Apr,df_Overview_May,df_Overview_Jun,df_Overview_Jul,df_Overview_Aug,df_Overview_Sep ])
df_Overview.head()
```

```
df = df_Overview.groupby(['Month','Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index'])['Abnormal_Return'].mean().reset_index()
df.head()
```

```
df=df.dropna()
df.info()
```

```
import plotly.express as px
fig = px.box(df_Overview, x="Month", y="Abnormal_Return", color="ESG_Index",
             notched=True, boxmode='group', # used notched shape
             title="Monthly Abnormal Return by ESG Rating",
            )
fig.show()
```

```
df_Agg=df.groupby(['Month','ESG_Index'], as_index=False)['Abnormal_Return'].agg(['mean','std']).reset_index()
df_Agg['error']=df_Agg['std']/2
df_Agg
```

```
fig = px.bar(df_Agg, x='Month', y='mean',color='ESG_Index',barmode='group',
             category_orders={"ESG_Index": ["Leader", "Laggard","Average"]})
fig.show()
```

```
###https://plotly.com/python/marker-style/

import plotly.express as px


fig1 = px.scatter(df_Agg, x='Month', y='mean',color='ESG_Index', marginal_y='box',category_orders={"ESG_Index": ["Leader", "Laggard","Average"]}, title='The Mean of Abnormal Return by ESG Index')
fig1.update_traces(marker=dict(size=12,
                              line=dict(width=2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
fig1.data[0].update(mode='markers+lines')

fig1.show()
```
