```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Jul= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Jul.csv")
df_Jul.head()
```

```
df_Jul=df_Jul.dropna()
df_Jul.head()
```

```
### formating data so that it can be calculated
df_Jul[['Size','Common Shares Outstdg','MktVal Co']]=df_Jul[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Jul.head()
```

```
Conds=[
       df_Jul['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Jul['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Jul['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Jul['ESG_Index'] =np.select(Conds, Choices)
df_Jul.head()
```

```
df_Jul[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Jul[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Jul.head()
```

```
### Generate the new variable:BM_Ratio
df_Jul['BM_Ratio'] = df_Jul['Book Value Per Shr Tangble']*df_Jul['Common Shares Outstdg']/df_Jul['MktVal Co']
df_Jul.head()
```

```
df_Jul.count()
```

```
### Generate the categories by quintile cut of size
df_Jul['Size_Index']=pd.qcut(df_Jul['Size'],5,labels=np.arange(1,6,1))
df_Jul.head()
```

```
df_Jul.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Jul['BM_Ratio_Index'] = df_Jul.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Jul.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Jul['Momentum_Index'] = df_Jul.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Jul.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Jul_Overview=df_Jul[['Symbol','Name','Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Jul_Overview.head()
```

```
df_Jul_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=True).size()
```

```
df_Jul_Overview_Benchmark = df_Jul_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Jul_Overview_Benchmark=df_Jul_Overview_Benchmark.rename(columns={"Monthly Return":"Benchmark_Return"})
df_Jul_Overview_Benchmark.head()
```

```
df_Jul_AR = pd.merge(df_Jul_Overview, df_Jul_Overview_Benchmark,  how='left', left_on=['Size_Index','BM_Ratio_Index','Momentum_Index'], right_on = ['Size_Index','BM_Ratio_Index','Momentum_Index'])
df_Jul_AR.head()
```

```
df_Jul_AR['Abnormal_Return']= df_Jul_AR['Monthly Return']-df_Jul_AR['Benchmark_Return']
df_Jul_AR.head()
```

```
###unpivot the table
df_Jul_AR['Month'] = "2020-07"
df_Jul_AR.head()
```
