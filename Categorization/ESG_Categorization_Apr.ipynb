```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Apr= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Apr.csv")
df_Apr.head()
```

```
df_Apr=df_Apr.dropna()
```

```
df_Apr=df_Apr.set_index('Symbol',drop=True)
df_Apr.head()
```

```
df_Apr.index.name=None
df_Apr.head()
```

```
### formating data so that it can be calculated
df_Apr[['Size','Common Shares Outstdg','MktVal Co']]=df_Apr[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Apr.head()
```

```
Conds=[
       df_Apr['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Apr['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Apr['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Apr['ESG_Index'] =np.select(Conds, Choices)
df_Apr.head()
```

```
df_Apr[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Apr[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Apr.head()
```

```
### Generate the new variable:BM_Ratio
df_Apr['BM_Ratio'] = df_Apr['Book Value Per Shr Tangble']*df_Apr['Common Shares Outstdg']/df_Apr['MktVal Co']
df_Apr.head()
```

```
df_Apr.count()
```

```
### Generate the categories by quintile cut of size
df_Apr['Size_Index']=pd.qcut(df_Apr['Size'],5,labels=np.arange(1,6,1))
df_Apr.head()
```

```
df_Apr.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Apr['BM_Ratio_Index'] = df_Apr.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Apr.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Apr['Momentum_Index'] = df_Apr.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Apr.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Apr_Overview=df_Apr[['Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Apr_Overview.head()
```

```
df_Apr_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).size()
```

```
df_Apr_Overview_Benchmark = df_Apr_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Apr_Overview_Benchmark.head()
```

```
df_Apr_Overview_Leader = df_Apr_Overview[df_Apr_Overview['ESG_Index']=='Leader']
df_Apr_Overview_Laggard = df_Apr_Overview[df_Apr_Overview['ESG_Index']=='Laggard']
df_Apr_Overview_Average = df_Apr_Overview[df_Apr_Overview['ESG_Index']=='Average']
df_Apr_Overview_Leader.head()
```

```
df_Apr_Overveiw_Leader_Mean = df_Apr_Overview_Leader.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Apr_Overveiw_Laggard_Mean = df_Apr_Overview_Laggard.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Apr_Overveiw_Average_Mean = df_Apr_Overview_Average.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Apr_Overveiw_Leader_Mean.head()
```

```
df_Apr_Overview_Benchmark['Leader Monthly Return']= df_Apr_Overveiw_Leader_Mean['Monthly Return']
df_Apr_Overview_Benchmark['Laggard Monthly Return']= df_Apr_Overveiw_Laggard_Mean['Monthly Return']
df_Apr_Overview_Benchmark['Average Monthly Return']= df_Apr_Overveiw_Average_Mean['Monthly Return']
df_Apr_Overview_Benchmark.head()
```

```
df_Apr_Overview_Benchmark['Leader Abnormal Return']=df_Apr_Overview_Benchmark['Leader Monthly Return']-df_Apr_Overview_Benchmark['Monthly Return']
df_Apr_Overview_Benchmark['Laggard Abnormal Return']=df_Apr_Overview_Benchmark['Laggard Monthly Return']-df_Apr_Overview_Benchmark['Monthly Return']
df_Apr_Overview_Benchmark['Average Abnormal Return']=df_Apr_Overview_Benchmark['Average Monthly Return']-df_Apr_Overview_Benchmark['Monthly Return']
df_Apr_Overview_Benchmark.head()
```

```
###unpivot the table
df_Apr_AR = df_Apr_Overview_Benchmark.melt(id_vars=['Size_Index', 'BM_Ratio_Index',	'Momentum_Index'], value_vars=['Leader Abnormal Return', 'Laggard Abnormal Return','Average Abnormal Return'],var_name='ESG Rating', value_name='Abnormal Return')
df_Apr_AR=df_Apr_AR.dropna()
df_Apr_AR.head()
```

```
df_Apr_AR['ESG Rating']=df_Apr_AR['ESG Rating'].apply(lambda x:'Leader' if x=='Leader Abnormal Return' else 'Laggard' if x=='Laggard Abnormal Return' else 'Average' )
df_Apr_AR.head()
```

```
df_Apr_AR['Month']='Apr'
df_Apr_AR.head()
```
