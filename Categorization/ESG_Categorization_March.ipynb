```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_March= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Mar.csv")
df_March.head()
```

```
df_March=df_March.dropna()
```

```
df_March=df_March.set_index('Symbol',drop=True)
df_March.head()
```

```
df_March.index.name=None
df_March.head()
```

```
### formating data so that it can be calculated
df_March[['Size','Common Shares Outstdg','MktVal Co']]=df_March[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_March.head()
```

```
df_March['ESG_Index'] = np.where(df_March['MSCI ESG Rtg Overall Company Rating'].isin(['A','AA','AAA']), 'High','Low')
df_March.head()
```

```
df_March[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_March[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_March.head()
```

```
### Generate the new variable:BM_Ratio
df_March['BM_Ratio'] = df_March['Book Value Per Shr Tangble']*df_March['Common Shares Outstdg']/df_March['MktVal Co']
df_March.head()
```

```
df_March.count()
```

```
### Generate the categories by quintile cut of size
df_March['Size_Index']=pd.qcut(df_March['Size'],5,labels=np.arange(1,6,1))
df_March.head()
```

```
df_March.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_March['BM_Ratio_Index'] = df_March.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_March.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_March['Momentum_Index'] = df_March.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_March.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_March_Overview=df_March[['Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
```

```
df_March_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).size()
```

```
df_March_Overview_Benchmark = df_March_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_March_Overview_Benchmark.head()
```

```
df_March_Overview_ESG_High = df_March_Overview[df_March_Overview['ESG_Index']=='High']
df_March_Overview_ESG_High.head()
```

```
df_March_Overveiw_ESG_High_Mean = df_March_Overview_ESG_High.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_March_Overveiw_ESG_High_Mean.head()
```

```
df_March_Overview_Benchmark['ESG_High_Return']= df_March_Overveiw_ESG_High_Mean['Monthly Return']
df_March_Overview_Benchmark.head()
```

```
df_March_Overview_Benchmark['Abnormal Return']=df_March_Overview_Benchmark['ESG_High_Return']-df_March_Overview_Benchmark['Monthly Return']
df_March_Overview_Benchmark.head()
```

```
df_March_Overview_Benchmark['Month']='March'
df_March_Overview_Benchmark.head()
```
