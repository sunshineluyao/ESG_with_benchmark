```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Feb= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Feb.csv")
df_Feb.head()
```

```
df_Feb=df_Feb.dropna()
```

```
df_Feb=df_Feb.set_index('Symbol',drop=True)
df_Feb.head()
```

```
df_Feb.index.name=None
df_Feb.head()
```

```
### formating data so that it can be calculated
df_Feb[['Size','Common Shares Outstdg','MktVal Co']]=df_Feb[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Feb.head()
```

```
Conds=[
       df_Feb['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Feb['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Feb['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Feb['ESG_Index'] =np.select(Conds, Choices)
df_Feb.head()
```

```
df_Feb[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Feb[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Feb.head()
```

```
### Generate the new variable:BM_Ratio
df_Feb['BM_Ratio'] = df_Feb['Book Value Per Shr Tangble']*df_Feb['Common Shares Outstdg']/df_Feb['MktVal Co']
df_Feb.head()
```

```
df_Feb.count()
```

```
### Generate the categories by quintile cut of size
df_Feb['Size_Index']=pd.qcut(df_Feb['Size'],5,labels=np.arange(1,6,1))
df_Feb.head()
```

```
df_Feb.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Feb['BM_Ratio_Index'] = df_Feb.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Feb.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Feb['Momentum_Index'] = df_Feb.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Feb.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Feb_Overview=df_Feb[['Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Feb_Overview.head()
```

```
df_Feb_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).size()
```

```
df_Feb_Overview_Benchmark = df_Feb_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Feb_Overview_Benchmark.head()
```

```
df_Feb_Overview_Leader = df_Feb_Overview[df_Feb_Overview['ESG_Index']=='Leader']
df_Feb_Overview_Laggard = df_Feb_Overview[df_Feb_Overview['ESG_Index']=='Laggard']
df_Feb_Overview_Average = df_Feb_Overview[df_Feb_Overview['ESG_Index']=='Average']
df_Feb_Overview_Leader.head()
```

```
df_Feb_Overveiw_Leader_Mean = df_Feb_Overview_Leader.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Feb_Overveiw_Laggard_Mean = df_Feb_Overview_Laggard.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Feb_Overveiw_Average_Mean = df_Feb_Overview_Average.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Feb_Overveiw_Leader_Mean.head()
```

```
df_Feb_Overview_Benchmark['Leader Monthly Return']= df_Feb_Overveiw_Leader_Mean['Monthly Return']
df_Feb_Overview_Benchmark['Laggard Monthly Return']= df_Feb_Overveiw_Laggard_Mean['Monthly Return']
df_Feb_Overview_Benchmark['Average Monthly Return']= df_Feb_Overveiw_Average_Mean['Monthly Return']
df_Feb_Overview_Benchmark.head()
```

```
df_Feb_Overview_Benchmark['Leader Abnormal Return']=df_Feb_Overview_Benchmark['Leader Monthly Return']-df_Feb_Overview_Benchmark['Monthly Return']
df_Feb_Overview_Benchmark['Laggard Abnormal Return']=df_Feb_Overview_Benchmark['Laggard Monthly Return']-df_Feb_Overview_Benchmark['Monthly Return']
df_Feb_Overview_Benchmark['Average Abnormal Return']=df_Feb_Overview_Benchmark['Average Monthly Return']-df_Feb_Overview_Benchmark['Monthly Return']
df_Feb_Overview_Benchmark.head()
```

```
###unpivot the table
df_Feb_AR = df_Feb_Overview_Benchmark.melt(id_vars=['Size_Index', 'BM_Ratio_Index',	'Momentum_Index'], value_vars=['Leader Abnormal Return', 'Laggard Abnormal Return','Average Abnormal Return'],var_name='ESG Rating', value_name='Abnormal Return')
df_Feb_AR=df_Feb_AR.dropna()
df_Feb_AR.head()
```

```
df_Feb_AR['ESG Rating']=df_Feb_AR['ESG Rating'].apply(lambda x:'Leader' if x=='Leader Abnormal Return' else 'Laggard' if x=='Laggard Abnormal Return' else 'Average' )
df_Feb_AR.head()
```

```
df_Feb_AR['Month']='Feb'
df_Feb_AR.head()
```
