```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Feb= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Feb.csv")
df_Feb.head()
```

```
df_Feb=df_Feb.dropna()
```

```
df_Feb=df_Feb.set_index('Symbol',drop=True)
df_Feb.head()
```

```
df_Feb.index.name=None
df_Feb.head()
```

```
### formating data so that it can be calculated
df_Feb[['Size','Common Shares Outstdg','MktVal Co']]=df_Feb[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Feb.head()
```

```
df_Feb[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Feb[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Feb.head()
```

```
### Generate the new variable:BM_Ratio
df_Feb['BM_Ratio'] = df_Feb['Book Value Per Shr Tangble']*df_Feb['Common Shares Outstdg']/df_Feb['MktVal Co']
df_Feb.head()
```

```
### Generate the categories by quintile cut and ESG rating
df_Feb['Size_Index']=pd.qcut(df_Feb['Size'],5,labels=np.arange(1,6,1))
df_Feb['Momentum_Index']=pd.qcut(df_Feb['Momentum'],5,labels=np.arange(1,6,1))
df_Feb['BM_Ratio_Index']=pd.qcut(df_Feb['BM_Ratio'],5,labels=np.arange(1,6,1))
df_Feb['ESG_Index']=np.where(df_Feb['MSCI ESG Rtg Overall Company Rating'].isin(['A','AA','AAA']), 'High','Low')
df_Feb.head()
```

```
df_Feb_Overview=df_Feb[['Size_Index','Momentum_Index','BM_Ratio_Index','ESG_Index','Monthly Return']]
```

```
df_Feb_Overview_Benchmark = df_Feb_Overview.groupby(by=['Size_Index','Momentum_Index','BM_Ratio_Index',],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Feb_Overview_Benchmark.head()
```

```
df_Feb_Overview_ESG_High = df_Feb_Overview[df_Feb_Overview['ESG_Index']=='High']
df_Feb_Overview_ESG_High.head()
```

```
df_Feb_Overveiw_ESG_High_Mean = df_Feb_Overview_ESG_High.groupby(by=['Size_Index','Momentum_Index','BM_Ratio_Index',],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Feb_Overveiw_ESG_High_Mean.head()
```

```
df_Feb_Overview_Benchmark['ESG_High_Return']= df_Feb_Overveiw_ESG_High_Mean['Monthly Return']
df_Feb_Overview_Benchmark.head()
```

```
df_Feb_Overview_Benchmark['Abnormal Return']=df_Feb_Overview_Benchmark['ESG_High_Return']-df_Feb_Overview_Benchmark['Monthly Return']
df_Feb_Overview_Benchmark.head()
```

```
df_Feb_Overview_Benchmark['Month']='Feb'
df_Feb_Overview_Benchmark.head()
```
