```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Aug= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Aug.csv")
df_Aug.head()
```

```
df_Aug=df_Aug.dropna()
df_Aug.head()
```

```
### formating data so that it can be calculated
df_Aug[['Size','Common Shares Outstdg','MktVal Co']]=df_Aug[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Aug.head()
```

```
Conds=[
       df_Aug['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Aug['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Aug['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Aug['ESG_Index'] =np.select(Conds, Choices)
df_Aug.head()
```

```
df_Aug[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Aug[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Aug.head()
```

```
### Generate the new variable:BM_Ratio
df_Aug['BM_Ratio'] = df_Aug['Book Value Per Shr Tangble']*df_Aug['Common Shares Outstdg']/df_Aug['MktVal Co']
df_Aug.head()
```

```
df_Aug.count()
```

```
### Generate the categories by quintile cut of size
df_Aug['Size_Index']=pd.qcut(df_Aug['Size'],5,labels=np.arange(1,6,1))
df_Aug.head()
```

```
df_Aug.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Aug['BM_Ratio_Index'] = df_Aug.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Aug.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Aug['Momentum_Index'] = df_Aug.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Aug.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Aug_Overview=df_Aug[['Symbol','Name','Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Aug_Overview.head()
```

```
df_Aug_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=True).size()
```

```
df_Aug_Overview_Benchmark = df_Aug_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Aug_Overview_Benchmark=df_Aug_Overview_Benchmark.rename(columns={"Monthly Return":"Benchmark_Return"})
df_Aug_Overview_Benchmark.head()
```

```
df_Aug_AR = pd.merge(df_Aug_Overview, df_Aug_Overview_Benchmark,  how='left', left_on=['Size_Index','BM_Ratio_Index','Momentum_Index'], right_on = ['Size_Index','BM_Ratio_Index','Momentum_Index'])
df_Aug_AR.head()
```

```
df_Aug_AR['Abnormal_Return']= df_Aug_AR['Monthly Return']-df_Aug_AR['Benchmark_Return']
df_Aug_AR.head()
```

```
###unpivot the table
df_Aug_AR['Month'] = "2020-08"
df_Aug_AR.head()
```
