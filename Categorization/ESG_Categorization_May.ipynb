```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_May= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_May.csv")
df_May.head()
```

```
df_May=df_May.dropna()
```

```
df_May=df_May.set_index('Symbol',drop=True)
df_May.head()
```

```
df_May.index.name=None
df_May.head()
```

```
### formating data so that it can be calculated
df_May[['Size','Common Shares Outstdg','MktVal Co']]=df_May[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_May.head()
```

```
Conds=[
       df_May['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_May['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_May['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_May['ESG_Index'] =np.select(Conds, Choices)
df_May.head()
```

```
df_May[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_May[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_May.head()
```

```
### Generate the new variable:BM_Ratio
df_May['BM_Ratio'] = df_May['Book Value Per Shr Tangble']*df_May['Common Shares Outstdg']/df_May['MktVal Co']
df_May.head()
```

```
df_May.count()
```

```
### Generate the categories by quintile cut of size
df_May['Size_Index']=pd.qcut(df_May['Size'],5,labels=np.arange(1,6,1))
df_May.head()
```

```
df_May.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_May['BM_Ratio_Index'] = df_May.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_May.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_May['Momentum_Index'] = df_May.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_May.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_May_Overview=df_May[['Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_May_Overview.head()
```

```
df_May_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).size()
```

```
df_May_Overview_Benchmark = df_May_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_May_Overview_Benchmark.head()
```

```
df_May_Overview_Leader = df_May_Overview[df_May_Overview['ESG_Index']=='Leader']
df_May_Overview_Laggard = df_May_Overview[df_May_Overview['ESG_Index']=='Laggard']
df_May_Overview_Average = df_May_Overview[df_May_Overview['ESG_Index']=='Average']
df_May_Overview_Leader.head()
```

```
df_May_Overveiw_Leader_Mean = df_May_Overview_Leader.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_May_Overveiw_Laggard_Mean = df_May_Overview_Laggard.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_May_Overveiw_Average_Mean = df_May_Overview_Average.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_May_Overveiw_Leader_Mean.head()
```

```
df_May_Overview_Benchmark['Leader Monthly Return']= df_May_Overveiw_Leader_Mean['Monthly Return']
df_May_Overview_Benchmark['Laggard Monthly Return']= df_May_Overveiw_Laggard_Mean['Monthly Return']
df_May_Overview_Benchmark['Average Monthly Return']= df_May_Overveiw_Average_Mean['Monthly Return']
df_May_Overview_Benchmark.head()
```

```
df_May_Overview_Benchmark['Leader Abnormal Return']=df_May_Overview_Benchmark['Leader Monthly Return']-df_May_Overview_Benchmark['Monthly Return']
df_May_Overview_Benchmark['Laggard Abnormal Return']=df_May_Overview_Benchmark['Laggard Monthly Return']-df_May_Overview_Benchmark['Monthly Return']
df_May_Overview_Benchmark['Average Abnormal Return']=df_May_Overview_Benchmark['Average Monthly Return']-df_May_Overview_Benchmark['Monthly Return']
df_May_Overview_Benchmark.head()
```

```
###unpivot the table
df_May_AR = df_May_Overview_Benchmark.melt(id_vars=['Size_Index', 'BM_Ratio_Index',	'Momentum_Index'], value_vars=['Leader Abnormal Return', 'Laggard Abnormal Return','Average Abnormal Return'],var_name='ESG Rating', value_name='Abnormal Return')
df_May_AR=df_May_AR.dropna()
df_May_AR.head()
```

```
df_May_AR['ESG Rating']=df_May_AR['ESG Rating'].apply(lambda x:'Leader' if x=='Leader Abnormal Return' else 'Laggard' if x=='Laggard Abnormal Return' else 'Average' )
df_May_AR.head()
```

```
df_May_AR['Month']='May'
df_May_AR.head()
```
