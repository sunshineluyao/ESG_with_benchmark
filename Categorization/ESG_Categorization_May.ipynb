```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_May= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_May.csv")
df_May.head()
```

```
df_May=df_May.dropna()
df_May.head()
```

```
### formating data so that it can be calculated
df_May[['Size','Common Shares Outstdg','MktVal Co']]=df_May[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_May.head()
```

```
Conds=[
       df_May['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_May['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_May['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_May['ESG_Index'] =np.select(Conds, Choices)
df_May.head()
```

```
df_May[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_May[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_May.head()
```

```
### Generate the new variable:BM_Ratio
df_May['BM_Ratio'] = df_May['Book Value Per Shr Tangble']*df_May['Common Shares Outstdg']/df_May['MktVal Co']
df_May.head()
```

```
df_May.count()
```

```
### Generate the categories by quintile cut of size
df_May['Size_Index']=pd.qcut(df_May['Size'],5,labels=np.arange(1,6,1))
df_May.head()
```

```
df_May.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_May['BM_Ratio_Index'] = df_May.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_May.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_May['Momentum_Index'] = df_May.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_May.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_May_Overview=df_May[['Symbol','Name','Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_May_Overview.head()
```

```
df_May_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=True).size()
```

```
df_May_Overview_Benchmark = df_May_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_May_Overview_Benchmark=df_May_Overview_Benchmark.rename(columns={"Monthly Return":"Benchmark_Return"})
df_May_Overview_Benchmark.head()
```

```
df_May_AR = pd.merge(df_May_Overview, df_May_Overview_Benchmark,  how='left', left_on=['Size_Index','BM_Ratio_Index','Momentum_Index'], right_on = ['Size_Index','BM_Ratio_Index','Momentum_Index'])
df_May_AR.head()
```

```
df_May_AR['Abnormal_Return']= df_May_AR['Monthly Return']-df_May_AR['Benchmark_Return']
df_May_AR.head()
```

```
###unpivot the table
df_May_AR['Month'] = "2020-05"
df_May_AR.head()
```
