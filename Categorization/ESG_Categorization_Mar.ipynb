```
### import packages
import pandas as pd
import numpy as np
### show all columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```
### Import data
df_Mar= pd.read_csv("https://raw.githubusercontent.com/sunshineluyao/ESG_with_benchmark/main/rawdata/SP1500_Mar.csv")
df_Mar.head()
```

```
df_Mar=df_Mar.dropna()
```

```
df_Mar=df_Mar.set_index('Symbol',drop=True)
df_Mar.head()
```

```
df_Mar.index.name=None
df_Mar.head()
```

```
### formating data so that it can be calculated
df_Mar[['Size','Common Shares Outstdg','MktVal Co']]=df_Mar[['Size','Common Shares Outstdg','MktVal Co']].apply(lambda x: x.str.replace(',', ''), axis=1)
df_Mar.head()
```

```
Conds=[
       df_Mar['MSCI ESG Rtg Overall Company Rating'].isin(['AAA','AA']),
       df_Mar['MSCI ESG Rtg Overall Company Rating'].isin(['CCC','B']),
       df_Mar['MSCI ESG Rtg Overall Company Rating'].isin(['BB','BBB','A'])
]
Choices = [
           'Leader',
           'Laggard',
           'Average'
]
df_Mar['ESG_Index'] =np.select(Conds, Choices)
df_Mar.head()
```

```
df_Mar[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']] = df_Mar[['Size', 'Book Value Per Shr Tangble','Common Shares Outstdg','MktVal Co','Momentum']].astype('float')
df_Mar.head()
```

```
### Generate the new variable:BM_Ratio
df_Mar['BM_Ratio'] = df_Mar['Book Value Per Shr Tangble']*df_Mar['Common Shares Outstdg']/df_Mar['MktVal Co']
df_Mar.head()
```

```
df_Mar.count()
```

```
### Generate the categories by quintile cut of size
df_Mar['Size_Index']=pd.qcut(df_Mar['Size'],5,labels=np.arange(1,6,1))
df_Mar.head()
```

```
df_Mar.groupby('Size_Index').size()
```

```
### generate the categories further by quintile of BM Ratio within each Size group. 
df_Mar['BM_Ratio_Index'] = df_Mar.groupby(['Size_Index'])['BM_Ratio'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Mar.groupby(by=['Size_Index','BM_Ratio_Index']).size()
```

```
### generate the categories further by quintile of Momentum within each previous group
df_Mar['Momentum_Index'] = df_Mar.groupby(by=['Size_Index','BM_Ratio_Index'])['Momentum'].transform(
                     lambda x: pd.qcut(x, 5, labels=np.arange(1,6,1)))
```

```
df_Mar.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index']).size()
```

```
df_Mar_Overview=df_Mar[['Size_Index','BM_Ratio_Index','Momentum_Index','ESG_Index','Monthly Return']]
df_Mar_Overview.head()
```

```
df_Mar_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).size()
```

```
df_Mar_Overview_Benchmark = df_Mar_Overview.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
```

```
df_Mar_Overview_Benchmark.head()
```

```
df_Mar_Overview_Leader = df_Mar_Overview[df_Mar_Overview['ESG_Index']=='Leader']
df_Mar_Overview_Laggard = df_Mar_Overview[df_Mar_Overview['ESG_Index']=='Laggard']
df_Mar_Overview_Average = df_Mar_Overview[df_Mar_Overview['ESG_Index']=='Average']
df_Mar_Overview_Leader.head()
```

```
df_Mar_Overveiw_Leader_Mean = df_Mar_Overview_Leader.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Mar_Overveiw_Laggard_Mean = df_Mar_Overview_Laggard.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Mar_Overveiw_Average_Mean = df_Mar_Overview_Average.groupby(by=['Size_Index','BM_Ratio_Index','Momentum_Index'],group_keys=True,sort=True,as_index=False).mean('Monthly Return')
df_Mar_Overveiw_Leader_Mean.head()
```

```
df_Mar_Overview_Benchmark['Leader Monthly Return']= df_Mar_Overveiw_Leader_Mean['Monthly Return']
df_Mar_Overview_Benchmark['Laggard Monthly Return']= df_Mar_Overveiw_Laggard_Mean['Monthly Return']
df_Mar_Overview_Benchmark['Average Monthly Return']= df_Mar_Overveiw_Average_Mean['Monthly Return']
df_Mar_Overview_Benchmark.head()
```

```
df_Mar_Overview_Benchmark['Leader Abnormal Return']=df_Mar_Overview_Benchmark['Leader Monthly Return']-df_Mar_Overview_Benchmark['Monthly Return']
df_Mar_Overview_Benchmark['Laggard Abnormal Return']=df_Mar_Overview_Benchmark['Laggard Monthly Return']-df_Mar_Overview_Benchmark['Monthly Return']
df_Mar_Overview_Benchmark['Average Abnormal Return']=df_Mar_Overview_Benchmark['Average Monthly Return']-df_Mar_Overview_Benchmark['Monthly Return']
df_Mar_Overview_Benchmark.head()
```

```
###unpivot the table
df_Mar_AR = df_Mar_Overview_Benchmark.melt(id_vars=['Size_Index', 'BM_Ratio_Index',	'Momentum_Index'], value_vars=['Leader Abnormal Return', 'Laggard Abnormal Return','Average Abnormal Return'],var_name='ESG Rating', value_name='Abnormal Return')
df_Mar_AR=df_Mar_AR.dropna()
df_Mar_AR.head()
```

```
df_Mar_AR['ESG Rating']=df_Mar_AR['ESG Rating'].apply(lambda x:'Leader' if x=='Leader Abnormal Return' else 'Laggard' if x=='Laggard Abnormal Return' else 'Average' )
df_Mar_AR.head()
```

```
df_Mar_AR['Month']='Mar'
df_Mar_AR.head()
```
